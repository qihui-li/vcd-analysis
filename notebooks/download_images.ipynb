### Download POPE images

```python
import json
import os
import requests

print("Check cwd:", os.getcwd())

pope_file = "./external/POPE/coco_pope_adversarial.json"
save_dir = "./data/images"
base_url = "http://images.cocodataset.org/val2014/"

data = []

with open(pope_file) as f:
    for line in f:
        data.append(json.loads(line))

images = set()
for item in data:
    images.add(item["image"])

downloaded = 0
for img in images:
    url = base_url + img
    path = os.path.join(save_dir, img)

    if not os.path.exists(path):
        print("Downloading", img)
        r = requests.get(url)
        with open(path, "wb") as f:
            f.write(r.content)

print("Downloaded", downloaded, "images in total")
```
